```python
complementos = []
for abrev in abreviacoes_df[0]:
    complementos.append(abrev)
    complementos.append(abrev.upper())
    complementos.append(abrev.lower())
    complementos.append(abrev.capitalize())
```

```python
complementos_df = pd.DataFrame(complementos).drop_duplicates().reset_index(drop=True)
complementos_df
```
