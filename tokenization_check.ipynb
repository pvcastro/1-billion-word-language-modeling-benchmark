```python
import pandas as pd
import spacy
from pathlib import Path

from scripts.generate_vocabulary import *
```

```python
path_in = 'full_harem_pt.csv'
path_out = 'full_harem_pt_sort.csv'
```

```python
data = pd.read_csv(path_in, header=None, escapechar='\\')
```

```python
data
```

```python
data_sorted = data.sort_values(by=0)
data_sorted
```

```python
data_sorted[0].astype(str)
```

```python
sorted_tokens = VocabularyTokenizer().process_all(data_sorted[0].astype(str))
```

```python
sorted_tokens[0]
```

```python
with Path(path_out).open(mode='w', encoding='utf8') as out:
    for tokens in sorted_tokens:
        out.write(str(tokens) + '\t' + str(len(tokens)) + '\n')
out.close()
```
