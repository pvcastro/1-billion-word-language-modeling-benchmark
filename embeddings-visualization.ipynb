```python
from allennlp.modules.elmo import Elmo, batch_to_ids
from allennlp.commands.elmo import ElmoEmbedder
import pandas as pd
import pprint
```

```python
wang_pd = pd.read_csv('/media/discoD/embeddings/wang2vec/wang2vec_s100_fixed.txt', skiprows=[0], header=None, sep=' ')
```

```python
options_file = "/media/discoD/models/elmo/options.json"
weight_file = "/media/discoD/models/elmo/elmo_pt_weights.hdf5"

elmo = ElmoEmbedder(options_file, weight_file, cuda_device=0)
tokens = ["Eu", "sou", "um", "estudante", "de", "mestrado", "."]
vectors = elmo.embed_sentence(tokens)

assert(len(vectors) == 3) # one for each layer in the ELMo output
assert(len(vectors[0]) == len(tokens)) # the vector elements correspond with the input tokens
```
